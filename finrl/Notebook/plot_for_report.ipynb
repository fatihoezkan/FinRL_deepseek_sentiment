```python
# load results from a file and print them in a table format
import argparse
import json
import os
import sys
import pandas as pd
import numpy as np


def load_data():
    df = pd.read_csv('/Users/fatihozkan/Desktop/finrl-deepseek-stock-prediction-fatih_clean/results.csv')
    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index('date').reset_index()
    df['A2C % Diff'] = 100 * (df["A2C Agent 2"] - df["A2C Agent 1"]) / df["A2C Agent 1"]
    df['SAC % Diff'] = 100 * (df["SAC Agent 2"] - df["SAC Agent 1"]) / df["SAC Agent 1"]
    df['PPO % Diff'] = 100 * (df["PPO Agent 2"] - df["PPO Agent 1"]) / df["PPO Agent 1"]
    df['TD3 % Diff'] = 100 * (df["TD3 Agent 2"] - df["TD3 Agent 1"]) / df["TD3 Agent 1"]
    df['A2C % Diff 3'] = 100 * (df["A2C Agent 3"] - df["A2C Agent 1"]) / df["A2C Agent 1"]
    df['SAC % Diff 3'] = 100 * (df["SAC Agent 3"] - df["SAC Agent 1"]) / df["SAC Agent 1"]
    df['PPO % Diff 3'] = 100 * (df["PPO Agent 3"] - df["PPO Agent 1"]) / df["PPO Agent 1"]
    df['TD3 % Diff 3'] = 100 * (df["TD3 Agent 3"] - df["TD3 Agent 1"]) / df["TD3 Agent 1"]
    return df

df = load_data()

df




```

```python
import matplotlib.pyplot as plt
# plot agent 1 and agent 2 results
df.plot(x='date', y=['A2C Agent 1', 'A2C Agent 2', 'SAC Agent 1', 'SAC Agent 2', 'PPO Agent 1', 'PPO Agent 2', 'TD3 Agent 1', 'TD3 Agent 2'], figsize=(10, 5), title='Agent 1 vs Agent 2 Results')

# plot agent 1 and agent 2 results only A2C and SAC
df.plot(x='date', y=['A2C Agent 1', 'A2C Agent 2', 'SAC Agent 1', 'SAC Agent 2'], figsize=(10, 5), title='Agent 1 vs Agent 2 Results (A2C and SAC)')
# plot agent 1 and agent 2 % difference
df.plot(x='date', y=['A2C % Diff', 'SAC % Diff'], figsize=(10, 5), title='Agent 1 vs Agent 2 % Difference (A2C and SAC)')


plt.show()



```
